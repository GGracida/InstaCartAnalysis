```python
# 4. create a new "max_order" column and group df by user_id and find the max orders for each user
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# create a new "loyalty_flag" column and assign loyalty label based on the amount of orders made
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# 5. group by loyalty_flag and obtain the mean, max and min of the prices column
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

###### The prices of productss purchased by loyal customers differ by  2  and 3 dollars less from the regular and new customers 

```python
# 6. create a new column "mean_prices_of_products" and group by user_id and find the mean prices for each user
ords_prods_merge['mean_prices_of_products'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# display the first 20 rows of the relevant columns 
ords_prods_merge[['user_id', 'mean_prices_of_products']].head(20)
```

```python
# create a new column "spending_flag" and assign spending label based on the given criteria.
ords_prods_merge.loc[ords_prods_merge['mean_prices_of_products'] >= 10, 'spending_flag'] = 'High spender'
ords_prods_merge.loc[ords_prods_merge['mean_prices_of_products'] < 10, 'spending_flag'] = 'Low Spender'
```

```python
#check the frequency of "spending_flag" column
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
# 7. create a new column "median_days_prior_order" and group by user_id and find the median of days_since_prior_order for each user
ords_prods_merge['median_days_prior_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# View relevant columns of updated dataframe
ords_prods_merge[['user_id', 'days_since_prior_order', 'median_days_prior_order']].head(20)
```

```python
# Mark all the values grater then 100 as nan
ords_prods_merge.loc[ords_prods_merge['prices'] >100, 'prices'] = np.nan
```

```python
# check the maximum value in the price column
ords_prods_merge['prices'].max()
```

```python
# create a new "frequency_flag" column and assign frequency label based on the given criteria
ords_prods_merge.loc[ords_prods_merge['median_days_prior_order'] > 20, 'frequency_flag'] = 'Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['median_days_prior_order'] <= 20) & (ords_prods_merge['median_days_prior_order'] > 10), 'frequency_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['median_days_prior_order'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
# check the frequency of "frequency_flag" column
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```
